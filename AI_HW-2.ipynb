```
class Node():
    
  def __init__(self, data):
    self.value = data
    self.left = self.right = None

class BinTree():
  
  def __init__(self, A):
    self.root = None
    if A:
      for element in A:
        self.insert(element)
  
  def insert(self, v):
    newNode = Node(v)
    if not self.root:
      self.root = newNode
    else:
      parentNode = self.root
      while True:
        if v > parentNode.value:
          if parentNode.right:
            parentNode = parentNode.right
          else:
            parentNode.right = newNode
            break
        else:
          if parentNode.left:
            parentNode = parentNode.left
          else:
            parentNode.left = newNode
            break
    
  def has(self, v):
    if not self.root:
      return False
    currentNode = self.root
    while True:
      if currentNode.value == v:
        return True
      if currentNode.value > v:
        if currentNode.left:
          currentNode = currentNode.left
        else:
          return False
      else:
        if currentNode.right:
          currentNode = currentNode.right
        else:
          return False

  def listHelper(self, node):
    if not node.left and not node.right:
      return [node.value]
    retList = []
    if node.left:
      retList += self.listHelper(node.left)
    retList.append(node.value)
    if node.right:
      retList += self.listHelper(node.right)
    return retList
      
  def get_ordered_list(self):
    if not self.root:
      return []   
    return self.listHelper(self.root)
  
  def clearHelper(self, node):
    if node.left:
      self.clearHelper(node.left)
    if node.right:
      self.clearHelper(node.right)
    node.left = node.right = None
  
  def clear(self):
    self.clearHelper(self.root)
    self.root = None
            
BT = BinTree([])
BT.insert(4)
BT.insert(9)
BT.insert(7)
BT.insert(1)
BT.insert(100)
BT.insert(5)
BT.insert(0)

print(BT.root.value)
print(BT.has(7))
print(BT.has(101))
print(BT.get_ordered_list())

Wood = BinTree([7,1,12,5,3])
print(Wood.get_ordered_list())

Wood.insert(6)
Wood.insert(0)

print(Wood.root.value)
print(Wood.has(7))
print(Wood.has(101))
print(Wood.get_ordered_list())

Wood.clear()

print(Wood.get_ordered_list())
print(Wood.root)
```
